```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot histograms for numerical columns
print("\nHistograms of Numerical Variables:")
df.hist(figsize=(15, 10), bins=20, color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Create boxplots for each numerical variable
plt.figure(figsize=(15, 10))
for idx, column in enumerate(df.columns, 1):
    plt.subplot(4, 4, idx)
    sns.boxplot(data=df[column], color='skyblue')
    plt.title(column)
    plt.tight_layout()
plt.show()
```

```python
# Pie chart for the proportion of NSP categories
nsp_counts = df['NSP'].value_counts()
plt.figure(figsize=(8, 6))
nsp_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'salmon'])
plt.title("Proportion of NSP Categories")
plt.ylabel("")
plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
print("\nCorrelation Heatmap:")
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pair plot for all variables
sns.pairplot(df)
plt.show()
```

```python
# Identify significant correlations
print("\nSignificant Correlations (Absolute Value > 0.7):")
high_corr = correlation_matrix[(correlation_matrix.abs() > 0.7) & (correlation_matrix.abs() < 1.0)]
print(high_corr.dropna(how='all').dropna(axis=1, how='all'))
```
